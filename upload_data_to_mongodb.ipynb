## Connect with MongoDB

```python
# Import the package
import pymongo
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Connect with the Mongodb Server
client = pymongo.MongoClient('localhost',27017)
```

## Set the name of the sheets

```python
# Create new database and connection
db = client['comp6235']
all_weather = db['all_weather']
london_historical_weather = db['london_historical_weather']
uk_energy = db['uk_energy']
```

## Insert the data into the database

```python
# Insert all weather data[last five years] into the database
file = pd.read_csv('./data/daily_weather.csv', chunksize=100000)
for chunk in file:
    data = chunk
    data = data[data['date']>'2013-01-01']
    data = data[data['date']<'2022-12-31']
    data_dict = data.to_dict(orient='records')
    all_weather.insert_many(data_dict)
```

```python
# Insert London Historical Weather Data into MongoDB for Weather Prediction
file2 = pd.read_csv('./data/london_weather.csv')
file2.replace(np.nan, 0, inplace=True)
data_dict2 = file2.to_dict(orient='records')
london_historical_weather.insert_many(data_dict2)
```

```python
# Insert UK Historical Weather Data into MongoDB for Weather Prediction
file3 = pd.read_csv('./data/uk_renewable_energy.csv')
file3.replace(np.nan, 0, inplace=True)
data_dict3 = file3.to_dict(orient='records')
uk_energy.insert_many(data_dict3)
```
